```python
import matplotlib.pyplot as plt
fb_views = [400, 500, 600, 800, 200]
yt_views = [800, 600, 900, 1500, 2000]
tw_views = [200, 300, 200, 1000, 1200]
days = range(1,6)
plt.plot(days, fb_views, label ="Facebook Views", color="b", marker = "D")
plt.plot(days, yt_views, label ="Youtube views", color="r", marker = "D")
plt.plot(days, tw_views, label ="Twitter", color="g", marker = "D")
plt.legend(loc = "upper left")
plt.title("Channel Views")
plt.grid(color='y', linestyle=":")
plt.savefig("Channel Views.png", dpi=400)
        
```
